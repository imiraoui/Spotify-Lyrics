## Downloading the data from the Spotify Web API

```python
#set the settings for the Spotify requests
with open("spotify_api.txt") as f:
    client_id, client_secret= f.readlines()
    client_id = client_id.strip("\n")
    redirect_uri = "http://localhost:8888/callback"
```

```python
import sys
import spotipy
import spotipy.util as util
import math
import pandas as pd

scope = 'user-library-read'

if len(sys.argv) > 1:
    username = sys.argv[1]
else:
    print (sys.argv[0],)
    sys.exit()

token = util.prompt_for_user_token(username, scope, client_id, client_secret, redirect_uri)

df_songs = pd.DataFrame(columns=["song_id","song_name","song_artist"])
if token:
    sp = spotipy.Spotify(auth=token)
    results = sp.current_user_saved_tracks(limit=50,offset=0)
    pages = math.ceil(results["total"]/50)
    for item in results['items']:
            track = item['track']
            track_id = track["id"]
            track_name = track['name']
            track_artist = track['artists'][0]['name']
            print (track['name'] + ' - ' + track['artists'][0]['name'])
            df_songs = df_songs.append(pd.Series([track_id,track_name,track_artist],index = df_songs.columns),ignore_index = True)
    for page in range(1,pages):
        results = sp.current_user_saved_tracks(limit=50,offset=page*50)
        print(page)
        for item in results['items']:
            track = item['track']
            track_id = track["id"]
            track_name = track['name']
            track_artist = track['artists'][0]['name']
            print (track['name'] + ' - ' + track['artists'][0]['name'])
            df_songs = df_songs.append(pd.Series([track_id,track_name,track_artist],index = df_songs.columns),ignore_index = True)
    df_songs.to_csv("songs_details.csv",index=False)
else:
    print ("Can't get token for", username)
```

```python
import sys
import spotipy
import spotipy.util as util

scope = 'user-library-read'

if len(sys.argv) > 1:
    username = sys.argv[1]
else:
    print (sys.argv[0],)
    sys.exit()
    
token = util.prompt_for_user_token(username, scope, client_id, client_secret, redirect_uri)
sp = spotipy.Spotify(auth=token)
for song in df_songs.index:
    if df_songs.loc[song,"searched_spotify"] == False:
        track = sp.track(df_songs.loc[song,"song_id"])
        df_songs.loc[song,"popularity"] = track["popularity"]
        df_songs.loc[song,"release_date"] = track["album"]["release_date"]
        df_songs.loc[song,"duration_ms"] = track["duration_ms"]
        df_songs.loc[song,"explicit"] = track["explicit"]
        df_songs.loc[song,"artists"] = str(track["artists"])
        audio_features = sp.audio_features(df_songs.loc[song,"song_id"])
        df_songs.loc[song,"danceability"] = audio_features[0]["danceability"]
        df_songs.loc[song,"energy"] = audio_features[0]["energy"]
        df_songs.loc[song,"key"] = audio_features[0]["key"]
        df_songs.loc[song,"loudness"] = audio_features[0]["loudness"]
        df_songs.loc[song,"mode"] = audio_features[0]["mode"]
        df_songs.loc[song,"speechiness"] = audio_features[0]["speechiness"]
        df_songs.loc[song,"acousticness"] = audio_features[0]["acousticness"]
        df_songs.loc[song,"instrumentalness"] = audio_features[0]["instrumentalness"]
        df_songs.loc[song,"liveness"] = audio_features[0]["liveness"]
        df_songs.loc[song,"valence"] = audio_features[0]["valence"]
        df_songs.loc[song,"tempo"] = audio_features[0]["tempo"]
        df_songs.loc[song,"searched_spotify"] = True
```

```python
for song in df_songs.index:
    if df_songs.loc[song,"searched_artists"] == False:
        artists = eval(df_songs.loc[song,"artists"])
        first_artist = artists[0]["id"]
        df_songs.loc[song,"first_artist"] = first_artist
        artist_results = sp.artist(first_artist)
        df_songs.loc[song,"artist_genres"] = str(artist_results["genres"])
        df_songs.loc[song,"artist_popularity"] = artist_results["popularity"]
        df_songs.loc[song,"searched_artists"] = True
```

## Getting the Data from Last.fm for the listening count

```python
import pylast

with open("lastfm_api.txt") as f:
    API_KEY, API_SECRET, username, password= f.readlines()
    API_KEY = API_KEY.strip("\n")
    API_SECRET = API_SECRET.strip("\n")
    username = username.strip("\n")
    password = password.strip("\n")
    
password_hash = pylast.md5(password)
network = pylast.LastFMNetwork(api_key=API_KEY, api_secret=API_SECRET,
                               username=username, password_hash=password_hash)

```

```python
df_songs["searched_lastfm"] = False
```

```python
for song in df_songs.index:
    if df_songs.loc[song,"searched_lastfm"] == False:
        try:
            track = network.get_track(df_songs.loc[song,"song_artist"], df_songs.loc[song,"song_name"])
            df_songs.loc[song,"lastfm_count"] = track.get_listener_count()
            df_songs.loc[song,"lastfm_wiki"] = track.get_wiki_content()
            df_songs.loc[song,"searched_lastfm"] = True
        except:
            df_songs.loc[song,"searched_lastfm"] = "Error"
            
```

## Let's use Selenium to parse the link to the lyrics and then the lyrics from AZlyrics

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

browser = webdriver.Chrome()
browser.get("http://azlyrics.com")
```

```python
df_songs["searched_lyrics"] = False
```

```python
for song in df_songs.index:
    search_txt = df_songs.loc[song,"song_name"] +" " + df_songs.loc[song,"song_artist"]
    browser.find_element_by_xpath("/html/body/div[2]/div[1]/div[1]/form/div/input").clear()
    browser.find_element_by_xpath("/html/body/div[2]/div[1]/div[1]/form/div/input").send_keys(search_txt)
    browser.find_element_by_xpath("/html/body/div[2]/div[1]/div[1]/form/div/input").send_keys(Keys.ENTER)
    try:
        lyrics_link = browser.find_element_by_xpath("/html/body/div[2]/div/div/div/table/tbody/tr/td/a").get_attribute("href")
        if "/lyrics/" in lyrics_link:
            df_songs.loc[song,"lyrics_link"] = lyrics_link
        else:
            df_songs.loc[song,"lyrics_link"] = browser.find_element_by_xpath("/html/body/div[2]/div/div/div/table/tbody/tr[2]/td/a").get_attribute("href")
        df_songs.loc[song,"searched_lyrics"] == True
        print(search_txt + " lyrics link added")
    except:
        print(search_txt + " has no lyrics")
        df_songs.loc[song,"searched_lyrics"] = True
```

```python
import time
for song in df_songs.index:
    if df_songs.loc[song,"lyrics"] == "Error" or df_songs.loc[song,"lyrics"].isnull() :
        time.sleep(1)
        browser.get(df_songs.loc[song,"lyrics_link"])
        try:
            df_songs.loc[song,"lyrics"] = browser.find_element_by_xpath("/html/body/div[3]/div/div[2]/div[5]").text
        except:
            df_songs.loc[song,"lyrics"] = "Error"
```

```python
import re
import string

def clean_lyrics(text):
    text = re.sub(r"\[(.*?)\]","",text)
    text = text.replace("\n"," ")
    text = text.replace("  "," ")
    text = text.lower()
    text = re.sub(r'[^\w\s]','',text)
    return text
```

```python
from langdetect import detect
#Cleaning the Lyrics and detecting the language
df_songs["lyrics_cleaned"] = df_songs.lyrics.apply(lambda x: clean_lyrics(x) if pd.notnull(x) else x)
df_songs["lang_lyrics"] = df_songs.lyrics.apply(lambda x: detect(x) if pd.notnull(x) else x)
```
